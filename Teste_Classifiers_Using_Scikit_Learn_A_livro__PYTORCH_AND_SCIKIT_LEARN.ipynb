4 A Tour of Machine Learning Classifiers Using Scikit-Learn
A nice feature in scikit-learn is that it allows us to readily visualize the decision tree model after training via the following code:

```python
from sklearn.datasets import load_iris
from sklearn import tree
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Train a decision tree classifier
tree_model = tree.DecisionTreeClassifier()
tree_model = tree_model.fit(X, y)

# Visualize the tree
feature_names = ['Sepal length', 'Sepal width',
'Petal length', 'Petal width']
tree.plot_tree(tree_model,
feature_names=feature_names,
filled=True)
plt.show()
```
